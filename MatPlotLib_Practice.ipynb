```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse, Circle, Rectangle
%config InlineBackend.figure_format = 'retina'
import seaborn as sns
```

```python
df = pd.read_csv('df.csv')
```

```python
yearavg = df.groupby('year').mean().reset_index()
temp = df.groupby(['year', 'team']).mean()
losing = temp.groupby('year').min().reset_index()
winning = temp.groupby('year').max().reset_index()
```

```python
# Initiate Plot
fig, ax = plt.subplots()
plt.rcParams['figure.figsize'] = (20,8)
sns.set_style('whitegrid', {"grid.linestyle": ":"})
ax.grid(which='major', axis='x')

# remove boarders except bottom
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('black')
ax.spines['bottom'].set_linestyle('dashed')

# Adjust tick on X-axis
ax.set_xticks(np.arange(1970, 2021, 5).tolist())

# lowest and highest point
plt.scatter(x = df.year, y = df.score, color = 'gray', alpha = 0.2)
plt.ylim(50,160)

# dotted line - average score for both teams
plt.plot(winning.year, winning.score, color = 'orange', lw = 3.2, marker = 'o')
plt.plot(losing.year, losing.score, color = 'steelblue', lw = 3.2, marker = 'o')

# title
plt.text(1983, 170, s = 'NBA Final  Games  Score', fontsize = '24', 
         fontfamily = 'serif', fontname  = 'American Typewriter'
)

# text and vline to illustrate scoreing ability for most Dominant teams during each time
plt.text(1980, 60, s = '  Showtime Lakers \n   Vs. Larry Bird', color = 'gray', fontsize = '18')
plt.vlines(1980, 60, 100, color = 'gray', lw = 2, alpha = 0.2)
plt.vlines(1988, 60, 100, color = 'gray', lw = 2, alpha = 0.2)

plt.text(1991, 138, s = '   Chicago Bulls', color = 'gray', fontsize = '18')
plt.vlines(1991, 97, 142, color = 'gray', lw = 2, alpha = 0.2)
plt.vlines(1998, 83, 142, color = 'gray', lw = 2, alpha = 0.2)

plt.text(2000, 133, s = '  Lakers', color = 'gray', fontsize = '18' )
plt.vlines(2000, 105, 136, color = 'gray', lw = 2, alpha = 0.2)
plt.vlines(2004, 85, 136, color = 'gray', lw = 2, alpha = 0.2)

plt.text(2014, 142, s = '   Warriors', color = 'gray', fontsize = '18' )
plt.vlines(2014, 99, 147, color = 'gray', lw = 2, alpha = 0.2)
plt.vlines(2019, 108, 147, color = 'gray', lw = 2, alpha = 0.2)

# text and circle to emphasise highest and lowest score of all time 
# Highest
ax.add_patch(Ellipse((1987, 141), 0.6, 3,
                     edgecolor='steelblue', lw=1, facecolor='none'))
plt.text(1986, 145, s = '141 Points', color = 'steelblue', fontsize = '24', 
         fontfamily = 'fantasy', fontname = 'Luminari', fontweight = 'bold')
plt.text(1986, 151, s = ' 1986 Lakers.\n  vs Celtics', fontsize = '15', fontweight = 'bold')

# Lowesst
ax.add_patch(Ellipse((1998, 54), 0.6, 3,
                     edgecolor='steelblue', lw=1, facecolor='none'))
plt.text(1997, 58, s = '54 Points', color = 'steelblue', fontsize = '24', 
         fontfamily = 'fantasy', fontname = 'Luminari', fontweight = 'bold')
plt.text(1997, 63, s = '1998 Jazz. vs Bulls', fontsize = '15', fontweight = 'bold')
ax.legend(['1st Average Score', '2nd Average Score'], frameon=False)
```
